```python
covid19 = COVID19Py.COVID19("https://coronavirus-tracker-api.herokuapp.com")
covid19 = COVID19Py.COVID19()
```

```python
locations = covid19.getLocations(timelines=True)
```

```python
def get_tot_cases(code_country):
    au = []
    for n in np.arange(0,len(locations),1):
        if locations[n]['country_code'] == code_country:
            au.append(locations[n])

    if len(au) != 1: 
        au_sum = []
        for i in np.arange(0,len(au),1):
            #print(i, conf[i])
            var = list(au[i]['timelines']['confirmed']['timeline'].values())
            for j in var:
                au_sum.append(j)

        tot_au_sum = []   
        tots_au_sum = []
        for i in np.arange(0,len(var),1):
            for j in np.arange(0,len(au),1):
                #print(i+j*68)
                tot_au_sum.append(au_sum[i+j*68])

        for i in np.arange(0,len(var),1):
            #print('{} : {}'.format(i*8, 7+8*i))
            tots_au_sum.append(sum(tot_au_sum[(i*len(au)):((len(au)-1)+len(au)*i)]))     
        return tots_au_sum
    else:
        return list(au[0]['timelines']['confirmed']['timeline'].values())
```

```python
conf = locations[0]['timelines']['confirmed']['timeline']
dates = []
for i in conf:
    #print(i, conf[i])
    dates.append(i[5:10])
```

```python
fig, ax = plt.subplots()

fig.tight_layout(rect=[0, 0.03, 2, 3])
font = {'size'   : 16}
plt.rc('font', **font)

ax.plot(dates, get_tot_cases('AU'))
ax.plot(dates, get_tot_cases('US'))
ax.plot(dates, get_tot_cases('IT'))
ax.plot(dates, get_tot_cases('CN'))
ax.plot(dates, get_tot_cases('ES'))
ax.plot(dates, get_tot_cases('BE'))
ax.plot(dates, get_tot_cases('CO'))
ax.plot(dates, get_tot_cases('EC'))
ax.plot(dates, get_tot_cases('AR'))

every_nth = 7
for n, label in enumerate(ax.xaxis.get_ticklabels()):
    if n % every_nth != 0:
        label.set_visible(False)
        
ax.set_xlabel('Date (m-d)')
ax.set_ylabel('Number of cases')
ax.legend(['Australia', 'United States', 'Italy', 'China','Spain','Belgium','Colombia','Ecuador','Argentina'])
fig.savefig('cases_over_time.png', dpi=300, facecolor='w', edgecolor='w',
       orientation='portrait', papertype=None, format=None,
       transparent=False, bbox_inches='tight', pad_inches=0.1)
```
